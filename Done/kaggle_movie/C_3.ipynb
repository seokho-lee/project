```python
import numpy as np
import pandas as pd
U_data = pd.read_csv('u.data', sep = "\t", header = None, encoding = 'utf')
U_data
U_data.rename(columns={0:'User_ID', 1:'Item_ID', 2:'rating', 3: 'timestemp'}, inplace=True)
U_data    # 첫번째 U_data (UserID, Item ID, rating, timestemp)
```

```python
U_Item = pd.read_csv('u.item', sep = "|", header = None, encoding = 'cp1252')
U_Item.rename(columns={0:'Item_ID', 1:'title', 2:'date', 5:'unknown',6: 'Action', 7: 'Adventure',8: 'Animation',
              9: 'Childrens', 10: 'Comedy', 11: 'Crime', 12:'Documentary', 13: 'Drama', 14:'Fantasy',
              15:'Film-Noir', 16: 'Horror', 17: 'Musical', 18: 'Mystery', 19: 'Romance', 20: 'Sci-Fi',
              21:'Thriller', 22: 'War', 23:'Western'}, inplace=True)
U_Item.drop([3,4], axis = 1, inplace = True)
U_Item         # 두번째 U_Item (Item_id, title, date, 장르 )
```

```python
U_user = pd.read_csv('u.user', sep = "|", header = None, encoding = 'utf')
U_user.rename(columns={0:'User_ID', 1:'age', 2:'sex', 3: 'occupation',4: 'zip'}, inplace=True)
U_user # 세번째 데이터(User_ID, age, sex, occupation, zip)
```

```python
young_age = U_user[(U_user['age'] >= 20) & (U_user['age'] < 40)] # 20 ~ 30 데이터
young_age
```

```python
U_u_U_d = young_age.merge(U_data, on = 'User_ID') # 20 ~ 30대 User_id로 묶음.
U_u_U_d
```

```python
occupation1 = U_u_U_d[(U_u_U_d['rating'] >= 4)]
occupation2 = occupation1.drop_duplicates(['User_ID', 'occupation'])
occupation2.groupby('occupation').count()['User_ID']
```

```python
df = U_u_U_d.merge(U_Item, on = 'Item_ID') # 20 ~ 30대 item_id 묶음.
df
```

```python
df1 = df[(df['rating'] >= 4) & (df['sex'] == 'M')] # 2030 남자가 rating을 높게 준 장르
df2 = df[(df['rating'] >= 4) & (df['sex'] == 'F')] # 2030 여자가 rating을 높게 준 장르
df5 = df[(df['rating'] >= 4)]
df3 = df1[['unknown','Action', 'Adventure','Animation',
              'Childrens', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
              'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi',
              'Thriller', 'War', 'Western']].sum()
df4 = df2[['unknown','Action', 'Adventure','Animation',
              'Childrens', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
              'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi',
              'Thriller', 'War', 'Western']].sum()
df6 = df5[['unknown','Action', 'Adventure','Animation',
              'Childrens', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
              'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi',
              'Thriller', 'War', 'Western']].sum()

df6 = df6.sort_values(ascending = False)
print(df6)

# mans_top = df3.nlargest(3)  # 20~30대 4점이상 남자가 선호한 장르 top 3
# womans_top = df4.nlargest(3)  # 20~30대 4점이상 여자가 선호한 장르 top 3
# total_top = df6.nlargest(3)  # 20~30대 4점이상 선호한 장르 top 3
# total_top
```

```python
# 장르별 직업 분포도
data_f = df.groupby('occupation').sum()
data_f.drop(['User_ID', 'age','Item_ID', 'rating', 'timestemp', 'unknown', 'Adventure','Animation',
              'Childrens', 'Crime', 'Documentary', 'Fantasy',
              'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi',
              'Thriller', 'War', 'Western'], axis = 1, inplace = True)
data_f
# data_f.max(axis = 1)['administrator']
```
