```python
data = df_data.copy()
data.columns = ['user id', 'item id', 'rating', 'timestamp']
item = df_item.copy()
item.columns = ["movie id", "movie title", "release date",
               "video release date", "IMDb URL", "unknown", "Action",
               "Adventure", "Animation", "Children's", "Comedy", "Crime",
               "Documentary", "Drama", "Fantasy", "Film-Noir", "Horror", "Musical",
               "Mystery", "Romance", "Sci-Fi", "Thriller", "War", "Western"]
user = df_user.copy()
user.columns = ['user id', 'age', 'gender', 'occupation', 'zip code']
```

```python
# data 중 시간을 알맞게 정리한 자료: 총 10만 개의 데이터
    # user id = 사용자 id
    # item id = 영화 id
    # rating = 평점
    # timestamp = 평점 올린 시간

from datetime import datetime
data = df_data.copy()
data.columns = ['user id', 'item id', 'rating', 'timestamp']
data['real_time'] = pd.to_datetime(data['timestamp'], unit = 's')
data

# 연 추출
data['year'] = data['real_time'].dt.year
data['year'] = data['year'].astype(str)
data

data['month'] = data['real_time'].dt.month

data['movie id'] = data['item id']
data
```

```python
data_user = pd.merge(left = data , right = user, how = "outer", on = "user id")
data_user_item = pd.merge(left = data_user , right = item, how = "outer", on = "movie id")
```

```python
data_user_item.info()
```

```python
data_user_item['year_release'] = data_user_item['release date'].str.split('-').str[-1]
data_user_item['month_release'] = data_user_item['release date'].str.split('-').str[-2]
```

```python
# 수치 범위 구간 직접 지정해서 레이블링하기
data_user_item['age_band'] = 0

data_user_item.loc[data_user_item['age']<10, 'age_band'] = '0s'
data_user_item.loc[(data_user_item['age']>=10) & (data_user_item['age']<20) , 'age_band'] = '10s'
data_user_item.loc[(data_user_item['age']>=20) & (data_user_item['age']<30) , 'age_band'] = '20s'
data_user_item.loc[(data_user_item['age']>=30) & (data_user_item['age']<40) , 'age_band'] = '30s'
data_user_item.loc[(data_user_item['age']>=40) & (data_user_item['age']<50) , 'age_band'] = '40s'
data_user_item.loc[(data_user_item['age']>=50) & (data_user_item['age']<60) , 'age_band'] = '50s'
data_user_item.loc[(data_user_item['age']>=60) & (data_user_item['age']<70) , 'age_band'] = '60s'
data_user_item.loc[(data_user_item['age']>=70) & (data_user_item['age']<80) , 'age_band'] = '70s'
data_user_item.loc[(data_user_item['age']>=80) & (data_user_item['age']<90) , 'age_band'] = '80s'
data_user_item.loc[(data_user_item['age']>=90) & (data_user_item['age']<100) , 'age_band'] = '90s'

data_user_item
```
