```python
fig = px.scatter(penguins, x="flipper_length_mm", y="bill_length_mm", color="species", symbol="sex")
fig.write_html("file.html")
fig.show()
```

```python
fig = px.box(penguins, x="flipper_length_mm", y="bill_length_mm")
fig.write_html("file.html")
fig.show()
# 각 mm마다의 분포를 나타내는 boxplot형태를 그려줌. outlier도 확인할 수 있다.
```

```python
fig = px.histogram(penguins, x="flipper_length_mm", color="species", kdf = True ,nbins = 30)
fig.show()
```

```python
fig = px.histogram(penguins, x="flipper_length_mm", color="species", marginal="box", nbins = 30)
# 종에 따른 펭귄 날개길이를 도수분포로 나타낸 것, 가장자리에 박스플롯을 그려준다.
fig.show()
```

```python
df = px.data.tips()
fig = px.box(df, y="tip", x="smoker", color="sex")
fig.show()
```

```python
df = px.data.tips()
fig = px.violin(df, y="tip", x="smoker", color="sex", hover_data=df.columns)
fig.show()
```

```python
df = px.data.tips()
fig = px.violin(df, y="tip", x="smoker", color="sex", box=True, hover_data=df.columns)
fig.show()
```

```python
import plotly.graph_objects as go
fig = go.Figure()

fig.add_trace(go.Violin(x=df['smoker'][ df['sex'] == 'Male' ],
                        y=df['tip'][  df['sex'] == 'Male'  ],
                        legendgroup='Yes', scalegroup='Yes', name='Male',
                        side='negative',
                        line_color='blue')
             )
fig.add_trace(go.Violin(x=df['smoker'][ df['sex'] == 'Female' ],
                        y=df['tip'][  df['sex'] == 'Female'  ],
                        legendgroup='No', scalegroup='No', name='Female',
                        side='positive',
                        line_color='green')
             )
fig.update_traces(meanline_visible=True)
fig.update_layout(violingap=0.5, violinmode='overlay')
fig.show()
```

```python
#pie chart: https://plotly.com/python/pie-charts/
import plotly.express as px
df = px.data.gapminder().query("year == 2007").query("continent == 'Europe'")
df.loc[df['pop'] < 2.e6, 'country'] = 'Other countries' # Represent only large countries
fig = px.pie(df, values='pop', names='country', title='Population of European continent')
fig.show()
```

```python
df = px.data.gapminder().query("year == 2007").query("continent == 'Europe'")
df.loc[df['pop'] < 5.e6, 'country'] = 'Other countries' # Represent only large countries
# 인구수 500만 미만의 국가들을 하나로 묶어 Other countries로 표기
```

```python
fig = px.pie(df, values='pop', names='country', title='Population of European continent')
# 'country'별 'pop' 비중을 pie chart 로 보여주줌
fig.show()
```
